```python
#import the modules
import os
import glob
import pandas as pd
```

## Uniting files from `general` directory

```python
#read the path
file_path = 'general'
```

```python
#list all the files from the directory
file_list = os.listdir(file_path)
file_list
```

```python
csv_files = glob.glob('{}/*{}.{}'.format('general','_general','csv'))
csv_files
```

```python
df_append = pd.DataFrame()
#append all files together
for file in csv_files:
            df_temp = pd.read_csv(file,sep=';')
            df_append = df_append.append(df_temp, ignore_index=True)
df_append
```

```python
df_append['webPublicationDateTime'] = df_append['webPublicationDate']
df_append['webPublicationDate'] = df_append['webPublicationDate'].str.split('T').str[0]
```

```python
df_append = df_append [[
    'id',
    'sectionName',
    'webPublicationDate',
    'webPublicationDateTime',
    'webTitle',
    'pillarId',
    'pillarName'
]]
```

```python
df_append
```

```python
df_append.to_csv('guardian-article-data_general.csv',index=False,sep=';')
```
